```
print("Unique values and their counts for 'Company_Type':")
print(df['Company_Type'].value_counts())
print("\nUnique values and their counts for 'Employee_Size':")
print(df['Employee_Size'].value_counts())
print("\nUnique values and their counts for 'Uses_AI':")
print(df['Uses_AI'].value_counts())
print("\nUnique values and their counts for 'Country' (top 10):")
print(df['Country'].value_counts().head(10))
print("\nUnique values and their counts for 'Industry' (top 10):")
print(df['Industry'].value_counts().head(10))
```

```
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['AI_ROI_Percent'], bins=20, kde=True)
plt.title('Distribution of AI ROI Percent')
plt.xlabel('AI ROI Percent')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['AI_Maturity_Score'], bins=20, kde=True)
plt.title('Distribution of AI Maturity Score')
plt.xlabel('AI Maturity Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='Company_Type', y='AI_ROI_Percent', data=df)
plt.title('AI ROI Percent by Company Type')
plt.xlabel('Company Type')
plt.ylabel('AI ROI Percent')

plt.subplot(1, 3, 2)
sns.boxplot(x='Employee_Size', y='AI_ROI_Percent', data=df)
plt.title('AI ROI Percent by Employee Size')
plt.xlabel('Employee Size')
plt.ylabel('AI ROI Percent')

plt.subplot(1, 3, 3)
sns.boxplot(x='Uses_AI', y='AI_ROI_Percent', data=df)
plt.title('AI ROI Percent by Uses AI')
plt.xlabel('Uses AI')
plt.ylabel('AI ROI Percent')

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='Company_Type', y='AI_Maturity_Score', data=df)
plt.title('AI Maturity Score by Company Type')
plt.xlabel('Company Type')
plt.ylabel('AI Maturity Score')

plt.subplot(1, 3, 2)
sns.boxplot(x='Employee_Size', y='AI_Maturity_Score', data=df)
plt.title('AI Maturity Score by Employee Size')
plt.xlabel('Employee Size')
plt.ylabel('AI Maturity Score')

plt.subplot(1, 3, 3)
sns.boxplot(x='Uses_AI', y='AI_Maturity_Score', data=df)
plt.title('AI Maturity Score by Uses AI')
plt.xlabel('Uses AI')
plt.ylabel('AI Maturity Score')

plt.tight_layout()
plt.show()
```

```
print("Top 10 AI Use Cases:")
print(df['Use_Case'].value_counts().head(10))
```

```
top_ai_use_cases = df[df['Use_Case'] != 'No Use Case']['Use_Case'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_ai_use_cases.index, y=top_ai_use_cases.values)
plt.title('Top 10 AI Use Cases (Excluding No Use Case)')
plt.xlabel('AI Use Case')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
numerical_cols = ['Year', 'Revenue_USD', 'AI_ROI_Percent', 'AI_Maturity_Score']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
